```python
def classify_review_baseline(text):
    text_lower = text.lower()
    predicted_tags = []
    
    bug_keywords = ["crash", "glitch", "stuck", "bug", "freeze", "break", "error", "broken"]
    
    feature_keywords = ["add", "please", "wish", "could", "needs", "support", "missing", "cant", "lack", "no mod"]
    
    praise_keywords = ["love", "great", "good", "amazing", "cozy", "perfect", "fun", "best", "blast", "gem"]
    
    analysis_keywords = ["cautious", "strategy", "mechanic"]

    if any(k in text_lower for k in bug_keywords):
        predicted_tags.append("Bug")
        
    if any(k in text_lower for k in feature_keywords):
        predicted_tags.append("Feature")
        
    if any(k in text_lower for k in praise_keywords):
        predicted_tags.append("Praise")

    if any(k in text_lower for k in analysis_keywords):
        predicted_tags.append("Analysis")

    return predicted_tags

print("Running Baseline ...")
df_real['baseline_pred'] = df_real['text'].apply(classify_review_baseline)
```

```python
candidate_labels = [
    "Bug Report (technical issues, glitches, crashes)", 
    "Feature Request (complaints about lack of content, suggestions, or missing mods)", 
    "Positive Praise (compliments, fun, immersion, or sarcasm)",
    "Game Analysis (neutral description of mechanics or gameplay loop)" 
]

def ai_classify_real_world(text):
    result = classifier(text, candidate_labels, multi_label=True)
    
    selected_tags = []
    for label, score in zip(result['labels'], result['scores']):
        if score > 0.3: 
            if "Bug Report" in label: selected_tags.append("Bug")
            elif "Feature Request" in label: selected_tags.append("Feature")
            elif "Positive Praise" in label: selected_tags.append("Praise")
            elif "Game Analysis" in label: selected_tags.append("Analysis")
            
    return selected_tags

print("Running AI ...")
df_real['ai_pred'] = df_real['text'].apply(ai_classify_real_world)
df_real[['text', 'tags', 'baseline_pred', 'ai_pred']]
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

# 1. Setup the Judge (Binarizer) on the REAL tags
mlb = MultiLabelBinarizer()
y_true = mlb.fit_transform(df_real['tags'])

# 2. Transform the predictions (Make sure to use df_real!)
y_baseline = mlb.transform(df_real['baseline_pred'])
y_ai = mlb.transform(df_real['ai_pred'])

# 3. Print the Text Report
print("\n--- BASELINE RESULTS (Real World) ---")
print(classification_report(y_true, y_baseline, target_names=mlb.classes_, zero_division=0))

print("\n--- AI PIPELINE RESULTS (Real World) ---")
print(classification_report(y_true, y_ai, target_names=mlb.classes_, zero_division=0))

# 4. (Optional) Visualizing the confusion helps specific analysis
# Since this is multi-label, we look at the specific class "Praise" or "Feature"
# Change "Praise" to "Bug" or "Feature" to see different graphs
target_label = "Praise" 
if target_label in mlb.classes_:
    class_id = list(mlb.classes_).index(target_label)
    disp = ConfusionMatrixDisplay.from_predictions(
        y_true[:, class_id], 
        y_ai[:, class_id],
        display_labels=[f"Not {target_label}", target_label],
        cmap=plt.cm.Blues
    )
    plt.title(f"AI Performance: Detecting '{target_label}'")
    plt.show()
```
